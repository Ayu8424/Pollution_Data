*Handling Missing Values and filling missing values*

```python
# Checking Null or NaN values 
pd.isnull(df).sum()
```

```python
# Dropping Column because they have missing values of more than 30/40%
df.drop(columns=['Xylene','NOx', 'Toluene', 'Benzene'], inplace=True)
```

```python
#Filling Missing Numerical Data
df.interpolate(method='linear', inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
#Dropping Duplicate if exist.
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

*Convert Columns to the correct type*

```python
#Convert the Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# We convert 'AQI_Bucket' to category to save memory, speed up operations, and improve data handling.
df['AQI_Bucket'] = df['AQI_Bucket'].astype('category')
```

```python
# Check how many unique cities are there
print(df['City'].nunique())  
```

*Add the Year and Month Column*

```python
#Create the Year column
df['Year'] = df['Date'].dt.year
```

```python
# Create the Month Column
df['Month'] = df['Date'].dt.month
```

```python
df.describe()
```

```python
# Cap Outliers Using IQR Method
Q1 = df["AQI"].quantile(0.25)
Q3 = df["AQI"].quantile(0.75)
IQR = Q3 - Q1

# Define upper and lower bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Cap the outliers (set max limit)
df["AQI"] = df["AQI"].apply(lambda x: upper_bound if x > upper_bound else (lower_bound if x < lower_bound else x))
```

```python
df = df[(df["AQI"] >= lower_bound) & (df["AQI"] <= upper_bound)]
```

*Distribution of AQI Levels*

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['AQI'], bins=30, kde=True, color="blue")
plt.xlabel("AQI Levels")
plt.ylabel("Frequency")
plt.title("Distribution of AQI Levels")
plt.show()
```

*Q1) How does pollution vary across different cities? (City-wise pollution level analysis)*

```python
city_pollution = df.groupby('City')['AQI'].mean().sort_values(ascending=False).round(0)
```

```python
city_pollution
```

```python
plt.figure(figsize=(22,7))
ax = sns.barplot(x= city_pollution.index, y= city_pollution.values, palette="coolwarm")
for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation=90)
plt.xlabel("City")
plt.ylabel("Average AQI")
plt.title("Average AQI by City (2015 - 2020)",fontdict={'fontsize': 18, 'fontweight': 'bold'})
plt.show()
```

*From the above graph, we can see that Delhi, Patna, Gurugram and Lucknow are the most polluted cities.*

```python
#Checking the TOP 10 Polluted Cities only.. 

# Group by city and compute the mean AQI
city_pollution_rank = df.groupby("City")["AQI"].mean().sort_values(ascending=False).round(0)

# Convert to DataFrame for better visualization
city_pollution_rank_df = city_pollution_rank.reset_index()

# Plot bar chart
plt.figure(figsize=(12,6))
ax = sns.barplot(data=city_pollution_rank_df.head(10), x="City", y="AQI")
plt.xticks(rotation=45)
for bars in ax.containers:
    ax.bar_label(bars)
plt.xlabel("City")
plt.ylabel("Average AQI")
plt.title("Top 10 Most Polluted Cities (Based on Average AQI)")
plt.show()
```

*Q4) What relationships exist between different pollutants? (Correlation analysis to identify key pollutant interactions)*

```python
# Correlation matrix
pollutants = ['PM2.5', 'PM10', 'NO', 'NO2', 'CO', 'SO2', 'O3']
corr_matrix = df[pollutants + ['AQI']].corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix[['AQI']].sort_values(by='AQI', ascending=False), 
            annot=True, cmap='coolwarm')
plt.title('Pollutant Correlation with AQI')
plt.show()
```

*Note: Based on the correlation analysis, we can conclude that PM2.5 has the strongest positive correlation with AQI (0.83), followed by PM10 (0.61). Other pollutants like NO2, NO, CO, SO2, and O3 show weaker correlations with AQI, with SO2 and O3 having the lowest correlation (0.24).*

```python
day_df = df.groupby(df['Date'].dt.date).agg({'AQI_Bucket': 'first'})  # or another appropriate aggregation

plt.figure(figsize=(10,6))
ax = sns.countplot(y='AQI_Bucket', data=day_df, order=df['AQI_Bucket'].value_counts().index)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('AQI Category Distribution')
plt.xlabel('Number of Days')
plt.ylabel('AQI Category')
plt.show()
```

*Note: From the above Graph we can say that there has been 649 days of Moderate, 867 days of Statisfactory and 211 day of Poor and very poor category and 277 days are good days and 5 are severe day*
